```python
import numpy as np
from mindspore import Parameter
from mindspore import Tensor, dtype
from mindspore import save_checkpoint

model_name = "BiT-M-R152x4"
file_name = "ms_name_152"
model = np.load(f"{model_name}.npz")
```

```python
block_key = []
no_block_key = []
for key in model.keys():
    if "gamma" in key:
        key = key.replace("gamma", "beta")
    elif "beta" in key:
        key = key.replace("beta", "gamma")
    if ('block' in key and "root" not in key) and ("proj" not in key):
        block_key.append(key)
    else:
        no_block_key.append(key)
ms_block_key = []
ms_no_block_key = []
with open(f'{file_name}.txt', 'r') as f:
    outs = f.readlines()
for out in outs:
    if "block" in out and "downsample" not in out:
        ms_block_key.append(out.strip())
    else:
        ms_no_block_key.append(out.strip())
A = list(range(len(no_block_key))) # no_block_key
B = [1,2,3,4,5,6,8,7,0] # ms_no_block_key
dict_map = {}
for a, b in zip(A, B):
    block_key.append(no_block_key[a])
    ms_block_key.append(ms_no_block_key[b])
for npz, ms in zip(block_key, ms_block_key):
    print(npz, ms)
```

```python
weights = []
prefix = "model."
for npz, ms in zip(block_key, ms_block_key):
    data_type = dtype.float32
    data = Tensor(model[npz], data_type)
    if model[npz].ndim == 4:
        data = data.transpose([3, 2, 0, 1])
    weight = Parameter(data, requires_grad=True)
    name = prefix + ms
    weights.append({"name": name, "data": weight})
    print(weights[-1]['name'])
save_checkpoint(weights, f"{model_name}.ckpt")
```
